```python
import pandas as pd
import numpy as np
from IPython.display import HTML
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,5))
frame1_text = f"In our first frame, we see the pass arrive to Cole Beasley in open space. The nearby defenders \n \
are players 8, 35, 38, and 98. Despite 8 being the nearest defender, \n \
the model gives almost no probability that he'll tackle due to his low speed and current orientation"
#fig.suptitle(frame1_text, fontsize=15)

base_dir = f"videos/video_2022100213_3554_frames"
ax.imshow(mpimg.imread(f"{base_dir}/frame0000000.png"))
ax.set_axis_off()

plt.figtext(0.5, 0.01, frame1_text, wrap=True, horizontalalignment='center', 
            bbox=dict(boxstyle="square", facecolor="beige"), fontsize=11)
plt.savefig("video_img_1_caption_box.png", bbox_inches='tight')

#plt.savefig('__', bbox_inches='tight')
```
